**MONTHLY SALES TREND**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'Amazon Sale Report.csv'
amazon_sale_report = pd.read_csv(file_path, encoding='ISO-8859-1')

# Convert the 'Date' column to datetime format, handling different year formats
amazon_sale_report['Date'] = pd.to_datetime(amazon_sale_report['Date'], format='%m-%d-%y', errors='coerce')
# The 'errors=coerce' argument tells pandas to handle invalid dates as NaT (Not a Time)

# Extract month and year from the 'Date' column
amazon_sale_report['Month'] = amazon_sale_report['Date'].dt.to_period('M')

# ... (rest of your code remains the same)
monthly_sales = amazon_sale_report.groupby('Month')['Amount'].sum().reset_index()

# Convert 'Month' to string for plotting
monthly_sales['Month'] = monthly_sales['Month'].astype(str) # Convert Period objects to strings for plotting

# Plot monthly sales trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Amount', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount (INR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Calculate category-wise sales
category_sales = amazon_sale_report.groupby('Category')['Amount'].sum().reset_index()

# Plot category-wise sales
plt.figure(figsize=(12, 6))
sns.barplot(data=category_sales, x='Category', y='Amount', palette='viridis')
plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount (INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the CSV file
file_path = 'Amazon Sale Report.csv'
amazon_sale_report = pd.read_csv(file_path, encoding='ISO-8859-1')

# Convert the 'Date' column to datetime format, handling different year formats
amazon_sale_report['Date'] = pd.to_datetime(amazon_sale_report['Date'], format='%m-%d-%y', errors='coerce')

# Investigate problematic dates
problematic_dates = amazon_sale_report[amazon_sale_report['Date'].isnull()]
print("Problematic dates:\n", problematic_dates[['Order ID', 'Date']])

# Extract month and year from the 'Date' column
amazon_sale_report['Month'] = amazon_sale_report['Date'].dt.to_period('M')

# Segment customers based on the number of orders they placed
customer_orders = amazon_sale_report['ship-postal-code'].value_counts().reset_index()
customer_orders.columns = ['Customer', 'Order Count']

# Plot customer order distribution
plt.figure(figsize=(12, 6))
sns.histplot(customer_orders['Order Count'], bins=30, kde=True)
plt.title('Customer Order Distribution')
plt.xlabel('Number of Orders')
plt.ylabel('Count of Customers')
plt.show()

# Segment customers based on their total spending
customer_spending = amazon_sale_report.groupby('ship-postal-code')['Amount'].sum().reset_index()
customer_spending.columns = ['Customer', 'Total Spending']

# Plot customer spending distribution
plt.figure(figsize=(12, 6))
sns.histplot(customer_spending['Total Spending'], bins=30, kde=True)
plt.title('Customer Spending Distribution')
plt.xlabel('Total Spending (INR)')
plt.ylabel('Count of Customers')
plt.show()

# Segment customers based on their location (state)
customer_location = amazon_sale_report['ship-state'].value_counts().reset_index()
customer_location.columns = ['State', 'Customer Count']

# Plot customer location distribution
plt.figure(figsize=(12, 6))
sns.barplot(data=customer_location, x='State', y='Customer Count', palette='viridis')
plt.title('Customer Location Distribution')
plt.xlabel('State')
plt.ylabel('Count of Customers')
plt.xticks(rotation=90)
plt.show()

# Combine factors to perform clustering
customer_data = amazon_sale_report.groupby('ship-postal-code').agg({
    'Amount': 'sum',
    'Order ID': 'count',
    'ship-state': 'first'
}).reset_index()
customer_data.columns = ['Customer', 'Total Spending', 'Order Count', 'State']

# Convert categorical state data to numeric
customer_data = pd.get_dummies(customer_data, columns=['State'], drop_first=True)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['Total Spending', 'Order Count'] + list(customer_data.columns[3:])])

# Perform KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Plot clusters
plt.figure(figsize=(12, 6))
sns.scatterplot(data=customer_data, x='Total Spending', y='Order Count', hue='Cluster', palette='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Total Spending (INR)')
plt.ylabel('Order Count')
plt.legend(title='Cluster')
plt.show()
```
